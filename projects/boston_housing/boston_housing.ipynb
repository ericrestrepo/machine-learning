### Implementation: Define a Performance Metric
It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. For this project, you will be calculating the [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination), R<sup>2</sup>, to quantify your model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the *mean* of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**. _A model can be given a negative R<sup>2</sup> as well, which indicates that the model is **arbitrarily worse** than one that always predicts the mean of the target variable._

For the `performance_metric` function in the code cell below, you will need to implement the following:
- Use `r2_score` from `sklearn.metrics` to perform a performance calculation between `y_true` and `y_predict`.
- Assign the performance score to the `score` variable.

```python
# TODO: Import 'r2_score'
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

### Question 7 - Grid Search
*What is the grid search technique and how it can be applied to optimize a learning algorithm?*

Grid search is a technique to allow you to tune a model and find the best hyper-parameter for a algorithm/estimator trained based upon a grid of values. Hyper-parameters are not directly learned in the algorithm, but instead are set when the estimator is constructed. Because of this basic construction as we go along in our analysis it can be tempting to tune the hyper-parameters increasingly until we could have a problem of overfitting without even realizing it. The grid search function is passed the following elements:

1. An estimator
2. A parameter space
3. A method for searching or sampling candidates
4. A cross-validation scheme
5. A scoring function

By using a grid search object, we can pass it an estimator, for example a DecisionTreeRegressor() and we can pass a parameter space as a dictionary with keys and values that are then vecorized and use dot product to create a matrix of combinations that are passed one by one to construct different estimators that are then fit to the data using the given cross-validation scheme, e.g. KFold, StratifiedKfold, ShuffleSplit, etc. and evaluated based upon the scoring function passed to grid search. This allows us to tune a hyper-parameter without overfitting the data. 

### Question 8 - Cross-Validation
*What is the k-fold cross-validation training technique? What benefit does this technique provide for grid search when optimizing a model?*  
**Hint:** Much like the reasoning behind having a testing set, what could go wrong with using grid search without a cross-validated set?

K-Fold cross-validation allows us to split up our data in to k folds or k bins of equal size. Instead of using a holdout technique just once, where we partition our data and save some to test our trained algorithm, with K-fold we can run k tests using each of the different bins as our test set. The training set for each test, would then be the combination of all of the other bins that were not used for the test. We then train our algorithm k times using a different bin as out test each time. We then get the average error across all of the trails instead of just with a normal holdout method. The train test split or hold out method always has the problem of wanting to maximize our training set to get the best classifier and maximizing our test set so we can most accurately measure how accurate our classifier performs. Also instead of just arbitrarily spliting our data into a train and test set, and having to randomly test and re-test to see which training set gets the best result, we can just run all of the training trials simultaneously.

Using a K-fold cross-validation allows grid search to find the most optimal hyper-parameter or set of hyper-parameters to fit the model or classifier, since you are more accurately randomizing and training and testing the classifier across the whole dataset and taking an average of the error, as opposed to arbitrarily breaking off a part of the dataset for training and leaving the rest for testing. If we only use a holdout method with grid search, then we run the risk of overfitting or underfitting our data since grid search seeks to find an optimal hyper-parameter, but it does not evaluate the cross validation method we pass to it. Any high variance or bias in our model based on how we split the data cannot be overcome by tuning the hyper-parameter(s).

For the `fit_model` function in the code cell below, you will need to implement the following:
- Use [`DecisionTreeRegressor`](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html) from `sklearn.tree` to create a decision tree regressor object.
  - Assign this object to the `'regressor'` variable.
- Create a dictionary for `'max_depth'` with the values from 1 to 10, and assign this to the `'params'` variable.
- Use [`make_scorer`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html) from `sklearn.metrics` to create a scoring function object.
  - Pass the `performance_metric` function as a parameter to the object.
  - Assign this scoring function to the `'scoring_fnc'` variable.
- Use [`GridSearchCV`](http://scikit-learn.org/0.17/modules/generated/sklearn.grid_search.GridSearchCV.html) from `sklearn.grid_search` to create a grid search object.
  - Pass the variables `'regressor'`, `'params'`, `'scoring_fnc'`, and `'cv_sets'` as parameters to the object. 
  - Assign the `GridSearchCV` object to the `'grid'` variable.

```python
# TODO: Import 'make_scorer', 'DecisionTreeRegressor', and 'GridSearchCV'
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import make_scorer
from sklearn.grid_search import GridSearchCV
from sklearn.model_selection import train_test_split

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    cv_sets = ShuffleSplit(X.shape[0], n_iter=10, test_size=0.20, random_state=0)
    cv = 
    # TODO: Create a decision tree regressor object
    regressor = DecisionTreeRegressor()

    # TODO: Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth': range(1,11)}

    # TODO: Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # TODO: Create the grid search object
    grid = GridSearchCV(estimator=regressor, param_grid=params, scoring=scoring_fnc, cv=cv_sets)
    
    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)
    
    # Return the optimal model after fitting the data
    return grid.best_estimator_
```
